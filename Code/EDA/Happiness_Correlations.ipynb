```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('/Users/axana/Ironhack/Projects/Happiness-Team/Data/happy_final.csv')
```

```python
dfnum = df._get_numeric_data()
dfnum.head()
```

```python
sns.pairplot(dfnum)
```

```python
df_after_2010 = df.loc[df["year"] > 2010]
df_after_2010
```

```python
dfnum_after_2010 = df_after_2010._get_numeric_data()
dfnum_after_2010.head()
```

```python
dfnum_after_2010.hist(bins = 15, figsize = (15, 6), layout = (2, 4));
```

```python
sns.pairplot(dfnum_after_2010)
```

```python
corr_data = dfnum_after_2010.drop(["year"], axis=1)
correlations_matrix = corr_data.corr()
correlations_matrix
```

```python
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```

```python
for column in corr_data:
    plt.figure()
    corr_data.boxplot([column])
```
